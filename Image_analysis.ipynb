```python
import cv2
# LOAD AN IMAGE USING 'IMREAD'
img = cv2.imread("/content/drive/MyDrive/Colab Notebooks/chihuahua.jpg")
# DISPLAY
cv2_imshow(img)
height,width = img.shape[:2]
```

##affine transform

```python
Tx,Ty =height/4,width/4
T = np.float32([[1,0,Tx],[0,1,Ty]])
result = cv2.warpAffine(img,T,(width,height))
cv2_imshow(result)
```

##rotate 

```python
height,width = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2),180,1)
result = cv2.warpAffine(img,rotation_matrix,(width,height))
cv2_imshow(result)
```

##scaling(use for sizing purpose)

```python
img_scaled = cv2.resize(img,None,fx=2,fy=2)
cv2_imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img,(300,300),interpolation = cv2.INTER_AREA)
cv2_imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img,(300,300),interpolation = cv2.INTER_CUBIC)
cv2_imshow(img_scaled)
```

```python
height,width = img.shape[:2]
start_row,start_col = int(height*0.35),int(width*0.35)
end_row,end_col = int(height*0.85),int(width*0.75)
crop_img = img[start_row:end_row,start_col:end_col]
cv2_imshow(img)
cv2_imshow(crop_img)
```
